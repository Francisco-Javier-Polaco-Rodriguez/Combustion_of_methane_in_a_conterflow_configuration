```python
from fluid_solver import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 32
L = 2
dx = L/N
x = np.linspace(0,L,N)
p,zero = np.meshgrid(x,x)
zero = zero * 0
```

```python
p_out = Gauss_Seidel(zero,zero,32,32,50000,np.sin(5*p),L/(N+1))
```

```python
def DDx(M:np.array,dx):
    I,J = M.shape
    ddxM = np.zeros((I,J))
    for j in range(1,J-1):
        ddxM[:,j] = (M[:,j+1]-2*M[:,j]+M[:,j-1])/dx**2
    return ddxM


X,Y = np.meshgrid(x,x)
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(X, Y,p_out)
```
